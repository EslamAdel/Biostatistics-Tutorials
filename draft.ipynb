```python
from itertools import product as prod
```

```python
s = {'H', 'M', 'L'}
p = {'H':0.01, 'M':0.04, 'L':0.95}
```

```python
space = list(prod(s, repeat=3))
```

```python
import pandas as pd
```

```python
import numpy as np 
```

```python
df = pd.DataFrame({"space":space})
```

```python
df['probs'] = df['space'].apply(lambda x: p[x[0]]*p[x[1]]*p[x[2]])
df
```

```python
df['X'] = df['space'].apply(lambda x: len([a for a in x if a == 'H']))
df['Y'] = df['space'].apply(lambda x: len([a for a in x if a == 'M']))
df
```

```python
df['XY']= df.apply(lambda x: str(x['X']) + str(x['Y']), axis = 1)
df
```

```python
df.groupby('XY', as_index= False)['probs'].sum()
```

```python
df.groupby('X', as_index= False)['probs'].sum()
```

```python
df[df['X'] == 1 ].groupby('Y')['probs'].sum()/df[df['X'] == 1 ]['probs'].sum()
```
